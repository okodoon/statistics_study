# 第七章　多次元の確率分布

## 共分散と相関係数

$$
V(X+Y) = V(X) + V(Y) + 2Cov(X,Y)
$$

ただし、$Cov(X,Y)$はX,Yの共分散と呼ばれ$E(X)=\mu_x$,$E(Y)=\mu_y$とすると、
$$
Cov(X,Y) = E{{((X-\mu_x)(Y-\mu_y))}}
$$

相関係数
$$
 \rho_{xy} = \frac{Cov(X,Y)}{\sqrt{V(X)V(Y)}}
$$

$Cov(X,Y)$はX,Yの相互作用、相互関連によるばらつき

X, Yの相互関連が正であるならば、X,Yを合わせたもののばらつきは大きくなる<br>
X, Yの相互関連が負であるならば、X,Yを合わせたもののばらつきは小さくなる

Xが上がった時にYも上がると（正の相関があると）X,Yの分散が大きくなる<br>
Xが上がった時にYが下がると（負の相関があると）X,Yの分散が小さくなる<br>
やじろべえみたいな感じ。リスク分散投資みたいな感じ


## 多次元正規分布

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm

#関数に投入するデータを作成
x = y = np.arange(-20, 20, 0.5)
X, Y = np.meshgrid(x, y)

z = np.c_[X.ravel(),Y.ravel()]

#二次元正規分布の確率密度を返す関数
def gaussian(x):
    #分散共分散行列の行列式
    det = np.linalg.det(sigma)
    print(det)
    #分散共分散行列の逆行列
    inv = np.linalg.inv(sigma)
    n = x.ndim
    print(inv)
    return np.exp(-np.diag((x - mu)@inv@(x - mu).T)/2.0) / (np.sqrt((2 * np.pi) ** n * det))

#2変数の平均値を指定
mu = np.array([0,0])
#2変数の分散共分散行列を指定
sigma = np.array([[1,0],[0,1]])

Z = gaussian(z)
shape = X.shape
Z = Z.reshape(shape)

#二次元正規分布をplot
fig = plt.figure(figsize = (15, 15))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm)
plt.show()
```
