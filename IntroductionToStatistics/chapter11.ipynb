# 第11章　推定

標本から母集団の分布を推定すること

## 点推定

$E(\bar{X}) = \mu$, $E(s^{2}) = \sigma^{2}$

$E(\bar{X})$や$E(s^{2})$を推定値と呼ぶ

### 点推定の手順：モーメント法
母集団分布が正規分布$N(\mu, \sigma^{2})$ の時の母集団の一次二次モーメント$\mu_1, \mu_2$を考える

$$
\hat{\mu} = \sum{X_i/n}
$$

$$
\hat{\sigma^{2}} = \sum{(X_i - \bar{X})^{2}/n}
$$

以上の式から母平均と母分散の推定量が求められる


モーメント法はk次元までの積率しか用いていないので不十分な場合がある（ヨクワカラン）。その問題点もカバーしているのが最尤法

### 点推定の手順：最尤法

最尤原理：「現実の標本は確率最大のものが実現した」という仮定

尤もらしさのことを尤度<br>
その関数を尤度関数という

最尤法：尤度関数を母数空間で最大にするものを推定値や推定量とするもの

最尤推定量$\hat{\theta}$が母数空間$\Theta$で最大になるのは<br>
同時確率分布の式が極大である時 => つまり微分して0になる => 傾きがないときが山の頂点なので<br>
その時の$\theta$が最尤


#### 例：ベルヌーイ分布の場合
$X_i=1$となる確率がp, $X_i=0$となる確率が1-pのベルヌーイ分布$B_i(1,p)$に従う時<br>
$r = \sum{X_i}$ とすると(ベルヌーイ分布とかだとコインの表が出た回数)<br>
尤度関数は
$$
L(p) = p^{r}(1-p)^{n-r}
$$
これをLog取って、L(p)をp微分して０になるときのpが<br>
1の相対頻度$\hat{p}=r/n$である

ベルヌーイ分布の1が出る確率の期待値がr/nなので、最尤値が一致したことを確認

### 推定の基準

#### 不偏性
nが小さい時にも大きい時にも、推定量の外れ具合が偏っていないこと


#### 一致性

標本が大きくなるにつれて、真の母数の値に近づく性質

「矢がどれだけ的に当たっているか」で説明される？


#### 漸近正規性
n => ∞の時の標本分布 ( = 漸近分布 ) が正規分布となる性質

#### 有効性

二つの推定量が、普遍性と一致性を満たしている場合に、分散が小さいものの方が相対的に好ましいこと


### 点推定の実例

#### 母集団が正規分布の場合

$$
\bar{X} = \sum_{i=1}^{n}{X_i/n}
$$
$$
S^2 = \sum_{i=1}^{n}{(X_i - \bar{X})^2/n}
$$
が最尤推定量となる

#### 母集団が二項分布の場合

母集団分布が標本$X_i=0,1$を出す母数pのベルヌーイ分布$B_i(1,p)$の二項分布であるとき

最尤推定量は
$$
\hat{p} = \bar{X}
$$
となる

#### 母集団がポアソン分布の場合

母集団分布が母数$\lambda$の$P_0(\lambda)$の時、

最尤推定量は
$$
\hat{\lambda} = \bar{X}
$$
となる

#### 母集団が一様分布の場合

母集団分布が区間(a,b)の一様分布(a<b)である時

$$
a = Min( X_1,X_2, ..., X_n )
$$

$$
b = Max( X_1,X_2, ..., X_n )
$$
が最尤推定量となる

#### ノンパラメトリックの場合

母集団分布の形がわからないので最尤法が使えない

モーメント法を使って求めるしかない

$$
\mu = \bar{X} = \sum_{i=1}^{n}{X_i/n}
$$

$$
S^2 = \sum_{i=1}^{n}{(X_i - \bar{X})^2/n}
$$
不偏推定量ではないため

$$
\sigma^2 = s^2 = \sum_{i=1}^{n}{(X_i - \bar{X})^2/(n-1)}
$$

## 区間推定

区間推定とは真の母数の値$\theta$が、ある区間(L,U)に入る確率を$1-\alpha$以上になるように保証する方法<br>
($\alpha$は$\theta$が区間に入らない確率)

$$
P(L \leqq \theta \leqq U) \geqq 1 - \alpha
$$

となる確率変数L,Uを求めること

L: 下側信頼限界<br>
U: 上側信頼限界

$1-\alpha$は信頼係数

区間[L,U]を$100(1-\alpha)$％の信頼区間と呼ぶ


### 母平均の信頼区間の推定

母平均$\mu$の信頼係数$1-\alpha$の信頼区間は

$$
[ \bar{X} - t_{\alpha/2}(n-1)・s/\sqrt{n}, \bar{X} + t_{\alpha/2}(n-1)・s/\sqrt{n} ]
$$
となる

### 母分散の信頼区間の推定

$(n-1)s^2/\sigma^2$は自由度n-1の$\chi^2$分布$\chi^2(n-1)$に従うから

$$
[\frac {(n-1)s^2}{\chi^2_{\alpha/2}(n-1)}, \frac {(n-1)s^2}{\chi^2_{1-\alpha/2}(n-1)} ]
$$
### 母平均の差の信頼区間

母集団分布が$N(\mu_1, \sigma_1^2), N(\mu_2, \sigma_2^2)$である二つの正規母集団から個別に二つの標本を抽出したとき

##### 二つの母集団の母分散が等しい場合

$$
[\bar{X} - \bar{Y} - t_{\alpha/2}(m+n-2)s\sqrt{\frac{1}{m}+\frac{1}{n}}, \bar{X} - \bar{Y} + t_{\alpha/2}(m+n-2)s\sqrt{\frac{1}{m}+\frac{1}{n}}]
$$

##### 二つの母集団の母分散が等しいと仮定できない場合

$$
[\bar{X} - \bar{Y} - t_{\alpha/2}(\nu^*)\sqrt{\frac{s_1^2}{m}+\frac{s_2^2}{n}}, \bar{X} - \bar{Y} + t_{\alpha/2}(\nu^*)\sqrt{\frac{s_1^2}{m}+\frac{s_2^2}{n}}]
$$

### 二項, ポアソン母集団の信頼区間
$B_i(1,p)$は

$$
\hat{p} = \bar{X}
$$
で点推定では推定されるが、pの信頼区間を求めることは難しい

そのため、中心極大定理を使って、標準正規分布で近似することでpの信頼区間を求める

$$
[\hat{p} - Z_{\alpha/2}\sqrt{\hat{p}(1-\hat{p})/n}, \hat{p} + Z_{\alpha/2}\sqrt{\hat{p}(1-\hat{p})/n}]
$$

##### ポアソン分布の場合

$$
[\hat{\lambda} - Z_{\alpha/2}\sqrt{\frac{\hat{\lambda}}{n}}, \hat{\lambda} + Z_{\alpha/2}\sqrt{\frac{\hat{\lambda}}{n}}]
$$

が$\lambda$の信頼区間
