# 第六章　確率分布



### 超幾何分布
N個からなる集団が2つの性質をもつときに、各集団がM個、N－M個であるとする。ここからn個を取り出すときに、<br>
M個の中からx個が取り出されるとすると、この確率分布は超幾何分布に従う。(非復元の場合である。復元の場合には二項分布)

$$
f(x) = \frac{{}_M\mathrm{C}_x\cdot{}_{N - M}\mathrm{C}_{n - x}}
{{}_N\mathrm{C}_n}
$$

### 二項分布・ベルヌーイ分布

#### ベルヌーイ試行
何かを行ったときに起こる結果が2つしかない試行をn回繰り返すこと。
例えば、コインの裏表など
2種類の可能な結果があり成功をS失敗をFとして、それらの確率をそれぞれ、p, 1-pとすると、Sがx回
Fがn-x回生じるとすれば、その確率は、
、
$$
f(x) = {}_n\mathrm{C}_xp^x(1 - p)^{n - x}
$$

この確率分布を`二項分布`という。<br>
つまり、ベルヌーイ試行をn回行って、成功する回数が従う確率分布。<br>
二項分布は、正規分布、ポアソン分布などを生み出す基礎的なものとして、確率分布の中心的な存在である。<br>
二項分布は
$$
Bi(n, p)
$$
で表す。また、
$$
Bi(1, p)$$
をベルヌーイ分布という。
確率変数Xが二項分布Bi(n,p)に従うならば、その期待値と分散は、
$$
E(X) = np
$$

$$
V(X) = np(1-p)
$$

### ポアソン分布

二項分布におけるnが大でpが小なケース

$f(x) = \frac{e^{-\lambda}\lambda^x}{x!}$

で表される.<br>
$\lambda$は期待値

申込の成約率が0.005 => 0.5%だとする<br>
1000件の申込みがあったとする

np = 5

### 幾何分布

ベルヌーイ試行を試行回数を決めずに繰り返した時、初めて成功Sが出現するするまでの試行回数Xが従う確率分布のこと。<br>
初めて成功Sが出現するまでの試行回数をx,確率をpとすると(q=1-p)

$$
f(x) = pq^{x-1}
$$

確率変数Xが幾何分布に従う時、その期待値、分散は、
$$
E(X) = \frac{1}{p}
$$

$$
V(X) = \frac{q}{p^2}
$$

と表すことができる。<br>
幾何分布は、時間を1,2,3と離散的に考えるとき、初めてのSを得る時間の長さの確率分布であり、離散的な待ち時間分布と呼ばれる。

###　正規分布
正規分布は代表的な連続型の確率分布。<br>
統計学における検定や推定、モデルの作成など様々な場面で活用される。<br>
正規分布に従う確率変数Xの確率密度関数f(x)は
$$
f(x) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp \left(-\frac{(x - \mu)^2}
{2\sigma^2} \right) \hspace{20px} (-\infty < x < \infty)
$$
確率変数Xが正規分布に従っている時、その期待値は、
$$
E(X)=\mu
$$
となる。<br>
つまり、期待値(平均値)そのものである。
<br>
分散は、
$$
V(X)=\sigma^2
$$
となり、σ^2は分散そのものである。<br>
このことから、平均をμ、分散をσ^2の正規分布といい、
$$
N(\mu, \sigma)
$$
で表す。

#### 中心極限定理

中心極限定理は、母集団分布が何であっても、抽出するサンプル数nが大きくなるにつれ標本平均の分布は、平均μ,分散$\frac{\sigma^2}{n}$の正規分布に近づく。

```python
# 統計学入門(東京大学出版会)の「中心極限定理」シミュレーション
# [0,1)の乱数を発生させ、0.1未満なら成功(1)、0.1以上なら失敗(0)とする。
# 繰り返し回数n(5,10,20,30,50,100)の二項乱数を250個生成し、標準化した確率分布が
# 正規分布に法則収束するかシミュレーションする。
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

n_list = [5, 10, 20, 30, 50, 100] # ベルヌーイ試行の回数(n)
N = 250 # 試行回数
p = 0.1 # ベルヌーイ試行の成功確率
z = np.array([]) # 標準化後の確率変数値の格納用リスト
E_list, V_list = [], [] # 平均値、分散の格納用リスト
E_t_list, V_t_list = [], [] # 期待値(理論値)、分散(理論値)の格納用リスト
x = {} # データ可視化用ディクショナリ

for n in n_list:
    r_n = np.random.rand(N, n) # 250×n の
    r_n = np.vectorize(lambda x: 1 if x<p else 0)(r_n)
    r = np.sum(r_n, axis=1)
    E = np.sum(r)/N # 平均値
    V = np.sum((r-r.mean())*(r-r.mean()))/N # 分散
    E_t = n*p # 期待値(理論値)
    V_t = n*p*(1-p) # 分散(理論値)
    #z = (r-E_t)/np.sqrt(V_t) # 標準化(理論値)
    z = (r-E)/np.sqrt(V) # 標準化

    E_list.append(E)
    E_t_list.append(E_t)
    V_list.append(V)
    V_t_list.append(V_t)
    x[n] = z
fig, axes = plt.subplots(3, 2, figsize=(10, 8), sharex=True, sharey=False)

# グラフ出力
for ax, n in zip(axes.ravel(), n_list):
    ax.set_title('Distribution after standardization - Bi('+str(n)+','+str(p)+')', fontsize = 10)
    ax.set_xlabel('Random variable(Z)')
    ax.set_ylabel('Frequency')
    ax.set_xlim(-4, 4)
    ax.grid()
    ax.hist(x[n], bins=40)

fig.suptitle('Central limit theorem - Simulation', fontsize = 14)
fig.tight_layout()
plt.show()

# 平均、分散出力
E_V = []
E_V.append(E_list)
E_V.append(E_t_list)
E_V.append(V_list)
E_V.append(V_t_list)
df = pd.DataFrame(E_V, columns=n_list, index=['平均','平均(理論値)','分散','分散(理論値)'
])
display(df)
```

### 指数分布

指数分布は、連続型確率分布の一つで、機械が故障してから次に故障するまでの期間や、災害が起こってから次に起こるまでの期間のように、次に何かが起こるまでの期間が従う分布。<br>
待ち時間分布の特性を持つ。<br>
ある期間に平均してλ回起こる現象が、次に起こるまでの期間が指数分布に従うとき、X=xとなる確率密度関数f(x)は次の式で表せる。

$$
f(x) = \begin{cases} \lambda e^{- \lambda x} & x \geq  0 \\ 
0 & x < 0\end{cases}
$$
この指数分布に従う確率変数Xの期待値と分散は、
$$
E(X) = \frac{1}{\lambda}
$$

$$
V(X) = \frac{1}{\lambda^2}
$$

### ガンマ分布

指数分布を一般化したもの

$x$: 確率変数<br>
$\alpha$: 形状母数<br>
$\lambda$: 尺度母数<br>

$$
f(x) = \frac{\lambda^{a}}{\Gamma(\alpha)}x^{\alpha - 1}e^{-\lambda x}
$$

任意の単位時間や単位面積 λ の間に 1 回起きることが期待できる出来事が、実際に起きるまでの時間の分布

指数分布は「期間 $\frac{1}{\lambda}$ に 1 回起きることが期待される事象が 1 回起きるまでの時間の分布」<br>
ガンマ分布は「期間$\frac{1}{\lambda}$に 1 回起こることが期待されるランダムな事象が $\alpha$回起こるまでの時間の分布」

#### ケーススタディ
10 年に一度の割合で故障するということは、形状母数 $\alpha$=1 、尺度母数 $\lambda$=10 ということになります<br>



### ベータ分布

確立的主観を表すのに使う<br>
ベイズにおける事前分布などが主なユースケース


$$
    f(x) = x^{\alpha - 1}(1-x)^{\beta - 1}/B(\alpha, \beta)
$$
