# 台紙・Figureの作成

台紙の作成にはplt.figure()を用いる。

```python
fig1 = plt.figure()
fig2 = plt.figure(figsize=(6.4, 4.8), dpi=100,facecolor='w', linewidth=0, edgecolor='w')
```

plt.figure()を実行すると、無地の台紙Figureが戻り値として返される。このときパラメータを入力すると図の設定ができる。

| 主な引数 | 説明 |
| --- | --- |
| figsize |	Figureのサイズ  横縦を(float | float)で指定。 |
| dpi | dpi。整数で指定。 |
| facecolor | 	図の背景色。Jupyterだと透過色になってたりする。 |
| linewidth |	図の外枠の太さ。デフォルトは0（枠なし）。 |
| edgecolor | 	図の枠の色。linewidthを指定しないと意味ない。 |
| subplotpars |	AxesSubplotの基準を指定する。 |
| tight_layout |	Trueにするとオブジェクトの配置が自動調整される。 |
| constrained_layout | Trueにするとオブジェクトの配置が自動調整される。 |

# グラフ(Axes)の追加

- Axesが1つ ⇒ Figure.add_subplot(111)（Figure.add_axes((0, 0, 1, 1))でも可）
- 同じ大きさのAxesを並べる ⇒ Figure.add_subplot()
- いろいろな大きさのAxesを並べる ⇒ GridSpec
- グラフの大きさ、位置をとにかく自由に決めたい ⇒ Figure.add_axes()

```python
ax1_1 = fig1.add_subplot(1, 1, 1)
ax1_2 = fig1.add_subplot(2, 1, 1)
ax2_1 = fig2.add_subplot(2, 1, 2)
```

**Figure.add_subplot(nrows, ncols, index)**<br>
無地の付箋Axes（正確にはAxesSubplotオブジェクト）が戻り値として返される。

- 縦nrows分割
- 横ncols分割
- index番目の位置に配置

| 主な引数 | 説明 |
| --- | --- |
| title | グラフのタイトル。 |
| facecolor | グラフの背景色。fcでも可。 |
| alpha | グラフの透明度を0～1で指定。 |
| zorder | オブジェクトが重なっていた時この値が大きい方が前面に描画される。 |
| xlabel | 横軸名。 |
| xmargin | データの最小値・最大値から横軸の最小値・最大値までのサイズ。 |
| xlim | 横軸の最小値・最大値を(float | float)で指定。 |
| xticks | 横軸の目盛線を表示する値をリストで指定。 |
| xticklabels | 横軸の目盛のラベルをリストで指定。 |
| sharex | 横軸を共有するAxesを指定。 |

```python
fig = plt.figure(facecolor='skyblue')
ax1 = fig.add_subplot(2, 3, 1)  # 2行3列の1番目
ax2 = fig.add_subplot(2, 2, 3)  # 2行2列の3番目
ax3 = fig.add_subplot(1, 4, 4)  # 1行4列の4番目

fig.show()
```

# plt.subplots()を使ってFigureとaxes両方を同時に作る

- plt.subplots(nrows, ncols)を実行すると、無地の台紙Figureと、無地の付箋Axes（複数の場合はそのリスト）が戻り値として返される。
- nrows，ncolsはAxesの配置を決めるパラメータで、指定するとFigureを縦nrows分割、横ncols分割した全ての位置のAxesの行列が返される。
- sharexをTrueにすると全ての横軸が共有される。sharexをcolにすると同じ列同士で横軸が共有される。
- このほかplt.figure()と同じようにパラメータを入力するとFigureの設定ができる。また、subplot_kw=dict()内にAxesのパラメータを指定できる

```python
fig, ax = plt.subplots()

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

fig, axes = plt.subplots(2, 2, sharey='row')
ax1 = axes[0, 0]
ax2 = axes[0, 1]
```

# 作ったAxesに対してプロットする

検証用データは西岡選手のテニスの試合の打点のデータとする

```python
import pandas as pd

# Business financial data: September 2021 quarter – CSV
df = pd.read_csv('../sample_data/shot-nishioka-nakashima.csv')
print(df.head())
```

### 折れ線グラフ

折れ線グラフはAxes.plot()。<br>
第一引数x（横軸の値）は省略可能。省略すると[0, 1, 2, ..., N-1]（x=range(len(y))）として扱われる。<br>
第三引数fmtには、color，marker，linestyleをひとまとめにして記述できる。<br>



```python
# 西岡選手が手前コートで打った打点のデータを作成する
df2 = df.query('bouncehit == "Hit" and hitplayer == "Nishioka" and bally > 12.5')

fig2 = plt.figure()
ax = fig2.add_subplot(111, xlabel=df.index.name, ylabel='number')

# 100データに限定
df2 = df2.iloc[1:100]
ax.plot(df2['ballx'])
ax.plot(df2['bally'],label='bally')

fig2.show()
```

### 散布図（Axes.scatter()）

```python
# 西岡選手が奥側コートで打った打点のデータを作成する
df3 = df.query('bouncehit == "Hit" and hitplayer == "Nishioka" and bally < 12.5')

fig3 = plt.figure()
ax = fig3.add_subplot(111, xlabel=df3['ballx'].name, ylabel=df3['bally'].name)

ax.scatter(df3['ballx'], df3['bally'], c='blue')

fig3.show()
```

ベースライン上で、左右満遍なく打っていることがわかる

### 棒グラフ（Axes.bar()）

```python

fig = plt.figure()
ax = fig.add_subplot(111, xlabel=df.index.name, ylabel='shot_count')

ax.bar(df.index, df['ballx'])

fig.show()
```

### ヒストグラム（Axes.hist()）

```python
# 打点のデータのコートx軸でのヒストグラム

fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111, xticks=range(0, 20, 2))
n, bins, patches = ax.hist(df['ballx'])

[ax.text((bins[i+1]+bins[i])/2, n[i], int(n[i]), horizontalalignment='center')
 for i in range(len(bins)-1)]

fig.show()
```
